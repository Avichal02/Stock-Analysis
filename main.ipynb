```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Indian

```python
tcs = pd.read_csv('Technology/tcs.csv')
tcs['Date'] = pd.to_datetime(tcs['Date'], infer_datetime_format=True)
tcs['Date'] = pd.to_datetime(tcs['Date']).dt.date
tcs = tcs.set_index('Date')
```

```python
gail = pd.read_csv('Oil and Gas/GAIL-2006-2010 - Sheet1.csv')
gail['Date'] = pd.to_datetime(gail['Date'], infer_datetime_format=True)
gail['Date'] = pd.to_datetime(gail['Date']).dt.date
gail = gail.set_index('Date')
```

```python
jsw = pd.read_csv('Constructions/jsw - sheet1.csv')
jsw['Date'] = pd.to_datetime(jsw['Date'], infer_datetime_format=True)
jsw['Date'] = pd.to_datetime(jsw['Date']).dt.date
jsw = jsw.set_index('Date')
```

```python
cipla = pd.read_csv('Pharmaceuticals/CIPLA-2006-2010 - Sheet1.csv')
cipla['Date'] = pd.to_datetime(cipla['Date'], infer_datetime_format=True)
cipla['Date'] = pd.to_datetime(cipla['Date']).dt.date
cipla = cipla.set_index('Date')
```

```python
hindu = pd.read_csv('FMCG/HINDUNILVR-2006-2010.csv')
hindu['Date'] = pd.to_datetime(hindu['Date'], infer_datetime_format=True)
hindu['Date'] = pd.to_datetime(hindu['Date']).dt.date
hindu = hindu.set_index('Date')
```

```python
sbi = pd.read_csv('Finance/SBI-2006-2010 - Sheet1.csv')
sbi['Date'] = pd.to_datetime(sbi['Date'], infer_datetime_format=True)
sbi['Date'] = pd.to_datetime(sbi['Date']).dt.date
sbi = sbi.set_index('Date')
```

```python
cipla['returns'] = (cipla['Close']/cipla['Close'].shift(1)) - 1
tcs['returns'] = (tcs['Close']/tcs['Close'].shift(1)) - 1
gail['returns'] = (gail['Close']/gail['Close'].shift(1)) - 1
jsw['returns'] = (jsw['Close']/jsw['Close'].shift(1)) - 1
sbi['returns'] = (sbi['Close']/sbi['Close'].shift(1)) - 1
hindu['returns'] = (hindu['Close']/hindu['Close'].shift(1)) - 1
```

```python
sbi['cumulative_returns'] = (1 + sbi['returns']).cumprod()
hindu['cumulative_returns'] = (1 + hindu['returns']).cumprod()
cipla['cumulative_returns'] = (1 + cipla['returns']).cumprod()
jsw['cumulative_returns'] = (1 + jsw['returns']).cumprod()
tcs['cumulative_returns'] = (1 + tcs['returns']).cumprod()
gail['cumulative_returns'] = (1 + gail['returns']).cumprod()
```

```python
sbi['cumulative_returns'].plot(label='SBI', figsize = (15, 7))
hindu['cumulative_returns'].plot(label='HINDULIVER')
cipla['cumulative_returns'].plot(label='CIPLA')
jsw['cumulative_returns'].plot(label='JSW')
gail['cumulative_returns'].plot(label='GAIL')
tcs['cumulative_returns'].plot(label='TCS')
plt.title('Cumulative Return v/s Time')
plt.legend()
plt.show()
```
