```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bharat = pd.read_csv('BHARATPETROLEUM-2006-2010 - Sheet1.csv')
hindustan = pd.read_csv('HINDUSTANPETROLEUM-2006-2010 - Sheet1.csv')
gail = pd.read_csv('GAIL-2006-2010 - Sheet1.csv')
indian = pd.read_csv('INDIANOIL-2006-2010 - Sheet1.csv')
```

```python
bharat['Date'] = pd.to_datetime(bharat['Date'], infer_datetime_format=True)
bharat['Date'] = pd.to_datetime(bharat['Date']).dt.date

hindustan['Date'] = pd.to_datetime(hindustan['Date'], infer_datetime_format=True)
hindustan['Date'] = pd.to_datetime(hindustan['Date']).dt.date

gail['Date'] = pd.to_datetime(gail['Date'], infer_datetime_format=True)
gail['Date'] = pd.to_datetime(gail['Date']).dt.date

indian['Date'] = pd.to_datetime(indian['Date'], infer_datetime_format=True)
indian['Date'] = pd.to_datetime(indian['Date']).dt.date
```

```python
bharat = bharat.set_index('Date')
hindustan = hindustan.set_index('Date')
gail = gail.set_index('Date')
indian = indian.set_index('Date')
```

```python
plt.figure(figsize=(15,7))
bharat['Close'].plot(label='BHARAT Closing Price', color = 'purple')
hindustan['Close'].plot(label='HINDUSTAN Closing Price', color = 'red')
gail['Close'].plot(label='GAIL Closing Price', color = 'blue')
indian['Close'].plot(label='INDIAN Closing Price', color ='skyblue')
plt.title("OIL and GAS Sector Stock Prices (2006-2010)")
plt.xlabel("Date")
plt.ylabel("Stock Price (in ₹)")
plt.legend()
```

```python
plt.figure(figsize=(15,7))
bharat['Volume'].plot(label='BHARAT', color = 'purple')
hindustan['Volume'].plot(label='HINDUSTAN', color = 'red')
gail['Volume'].plot(label='GAIL', color = 'blue')
indian['Volume'].plot(label='INDIAN', color ='skyblue')
plt.title("Volume Traded by BHARAT, HINDUSTAN, GAIL, INDIAN (2006-2010)")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Volume Traded')
```

```python
bharat['Total Traded'] = bharat['Open'] * bharat['Volume']
hindustan['Total Traded'] = hindustan['Open'] * hindustan['Volume']
gail['Total Traded'] = gail['Open'] * gail['Volume']
indian['Total Traded'] = indian['Open'] * indian['Volume']
```

```python
plt.figure(figsize = (15,7))
bharat['Total Traded'].plot(label = 'BHARAT Market Cap')
hindustan['Total Traded'].plot(label = 'HINDUSTAN Market Cap')
gail['Total Traded'].plot(label = 'GAIL Market Cap')
indian['Total Traded'].plot(label = 'INDIAN Market Cap')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Market Cap')
plt.title('Market Capitalisation of BHARAT, HINDUSTAN, GAIL, INDIAN  (2006-2010)')
```

```python
bharat['returns'] = (bharat['Close']/bharat['Close'].shift(1)) - 1
hindustan['returns'] = (hindustan['Close']/hindustan['Close'].shift(1)) - 1
gail['returns'] = (gail['Close']/gail['Close'].shift(1)) - 1
indian['returns'] = (indian['Close']/indian['Close'].shift(1)) - 1
```

```python
bharat['returns'].hist(bins = 70, label = 'BHARAT', alpha = 0.5, figsize = (13, 6))
indian['returns'].hist(bins = 70, label = 'HINDUSTAN', alpha = 0.5)
gail['returns'].hist(bins = 70, label = 'GAIL', alpha = 0.5)
hindustan['returns'].hist(bins = 70, label = 'INDIAN', alpha = 0.5)
plt.legend()
```

```python
bharat['returns'].plot(kind='kde', label = 'BHARAT', figsize = (15, 6))
hindustan['returns'].plot(kind='kde', label = 'HINDUSTAN')
gail['returns'].plot(kind='kde', label = 'GAIL')
indian['returns'].plot(kind='kde', label = 'INDIAN')
plt.legend()
```

```python
box_df =  pd.concat([bharat['returns'], hindustan['returns'], gail['returns'], indian['returns']], axis = 1)
box_df.columns = ['BHARAT Returns', 'HINDUSTAN Returns', 'GAIL Returns', 'INDIAN Returns']
box_df.plot(kind = 'box', figsize = (16,6))
```

```python
scatter_matrix(box_df, figsize = (10, 10), hist_kwds={'bins':50}, alpha = 0.25)
plt.show()
```

```python
bharat['cumulative_returns'] = (1 + bharat['returns']).cumprod()
hindustan['cumulative_returns'] = (1 + hindustan['returns']).cumprod()
gail['cumulative_returns'] = (1 + gail['returns']).cumprod()
indian['cumulative_returns'] = (1 + indian['returns']).cumprod()
```

```python
bharat['cumulative_returns'].plot(label='BHARAT', figsize = (15, 7))
hindustan['cumulative_returns'].plot(label='HINDUSTAN')
gail['cumulative_returns'].plot(label='GAIL')
indian['cumulative_returns'].plot(label='INDIAN')
plt.title('Cumulative Return v/s Time')
plt.legend()
plt.show()
```
